```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x)
```

```python
x = np.arange(-10, 10, 0.001)
y = f(x)
y
```

```python
plt.figure()
```

```python
def f2(f, x):
    h = 1e-7
    return (f(x + h) - f(x)) / h
```

```python
plt.plot(x,f2(f,x))
plt.plot(x, y)
plt.title('y=sin(x)')
plt.xlabel('x')
plt.ylabel('y=sin(x)')
plt.show()
```
